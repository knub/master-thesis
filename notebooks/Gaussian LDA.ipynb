```python
def plot_normal(mean, cov):
    points = np.random.multivariate_normal(mean, cov, 2500)
    plt.figure()
    plt.scatter(points[:,0], points[:,1])
    plt.xlim([-10, 10])
    plt.ylim([-10, 10])
    
    
mean = np.array([0.0, 0.0])
cov = np.array([[1.00, 0.9],
                [0.1, 1.0]])
plot_normal(mean, cov)
```

```python
scale_matrix = np.array([[10.0, 0.0],
                         [0.0, 1.0]])
cov = scipy.stats.invwishart.rvs(df=10, scale=scale_matrix, size=1)

plot_normal(mean, cov)
```
